```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/error_logs.csv")
```

```python
df.info()
```

```python
df["effort_level"].value_counts().plot(kind="bar")
plt.title("Distribution of Debug Effort Levels")
plt.xlabel("Effort Level")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["time_to_fix"], bins=10, kde=True)
plt.title("Time to Fix Distribution")
plt.xlabel("Minutes")
plt.show()
```

```python
sns.countplot(data=df, x="error_type", hue="effort_level")
plt.title("Error Type vs Debug Effort")
plt.xticks(rotation=30)
plt.show()
```

```python
numeric_df = df[["line_number", "time_to_fix"]]
numeric_df.corr()
```

### EDA Summary
- Most errors fall under Medium effort level.
- Syntax and Import errors usually require low effort.
- Runtime and recursion-related errors tend to require higher effort.
- Time to fix shows right-skewed distribution indicating few complex bugs.
